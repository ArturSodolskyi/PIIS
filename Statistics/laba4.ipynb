```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("statistics.csv")
```

```python
x = np.array(data['algorithm']).reshape(-1, 1)
y = np.array(data.drop('algorithm', axis=1))
```

```python
regr = LinearRegression()
regr.fit(x, y)
```

```python
all_algorithms = [[0], [1], [2], [3], [4], [5]]
prediction = regr.predict(all_algorithms)
pred_array = []
for i in range(len(prediction)):
  pred = prediction[i]
  pred_array.append([all_algorithms[i][0], round(pred[0]), pred[1], round(pred[2])])
next_games = pd.DataFrame(pred_array, columns=['algorithm', 'win', 'time', 'score'])
next_games
```

```python
next_games.to_csv('nextGamesForecasting.csv')
```
